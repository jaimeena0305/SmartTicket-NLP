```
import pandas as pd
import random

# Define categories and templates for synthetic data
categories = {
    "Hardware": [
        "My mouse is not working",
        "The printer is jamming",
        "Screen is flickering",
        "Laptop won't turn on",
        "Need a new monitor",
        "Keyboard keys are stuck",
        "Battery drains too fast"
    ],
    "Software": [
        "Excel keeps crashing",
        "Need to install Python",
        "Adobe Reader update failed",
        "Cannot open this PDF",
        "Zoom is freezing",
        "Browser cache needs clearing",
        "Application throws error 404"
    ],
    "Network": [
        "Wifi is very slow",
        "Cannot connect to VPN",
        "Internet is down",
        "Packet loss in video calls",
        "DNS resolution error",
        "Ethernet cable not detected",
        "Can't access internal portal"
    ],
    "Access": [
        "Forgot my password",
        "Locked out of account",
        "Need access to HR folder",
        "Badge not working",
        "Reset MFA token",
        "Permission denied for shared drive",
        "Login timeout error"
    ]
}

urgency_keywords = ["urgent", "asap", "immediately", "critical", "blocking", "help"]

def generate_ticket():
    category = random.choice(list(categories.keys()))
    base_text = random.choice(categories[category])

    # Randomly add urgency
    if random.random() > 0.7:
        urgency_text = f"This is {random.choice(urgency_keywords)}."
        text = f"{base_text}. {urgency_text}"
    else:
        text = base_text

    return text, category

# Generate 1000 synthetic tickets
data = []
for _ in range(1000):
    text, label = generate_ticket()
    data.append([text, label])

df = pd.DataFrame(data, columns=["text", "category"])

# Save to CSV
df.to_csv("ticket_data.csv", index=False)
print("âœ… generated ticket_data.csv with 1000 samples.")
```
